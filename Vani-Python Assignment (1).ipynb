3. What are pre-defined keywords in Python?

  6. def/class/with/as/lambda >> define functions


4. Can keywords be used as variable names?
 * No, keywords cannot be used as variable names in Python. Keywords are reserved words that have predefined meanings and functions within the Python language. Using a keyword as a variable name would create a conflict and result in a syntax error because Python would not be able to distinguish between its reserved function and your intended variable usage.


5. What is mutability in Python?
  * Mutability refers to whether an object in Python can be changed after it has been created. Python classifies objects into two types based on their mutability:

   1. Mutable Objects:
 These objects can be altered in place, meaning their content or state can be changed without creating a new object.
Ex: list, dict, set, bytearray

  2. Immutable Objects
These objects cannot be changed once they are created. Any modification to an immutable object creates a new object.
Ex: int, float, bool, str, tuple, frozenset

6.  Why are lists mutable, but tuples are immutable?
   * Tuples and lists are the same in every way except two: tuples use parentheses instead of square brackets, and the items in tuples cannot be modified but the items in lists can be modified. We often call lists mutable meaning they can be changed and tuples immutable meaning they cannot be changed.


7. What is the difference between “==” and “is” operators in Python?
   * The key difference between the "==" and "is" operators in Python lies in what they compare:
     "==" equality operator: It checks if the values of two objects are the same.
     "is" identity operator: It checks if two variables refer to the same object in memory.
In simpler terms, "==" compares the contents, while "is" compares the memory addresses.



8. What are logical operators in Python?
   * Logical operators in Python are used to combine or modify conditional statements. They work with boolean values "True or False" and return a boolean result. There are three main logical operators in Python: and, or, and not.
 1. and: Returns True if both operands are True, otherwise returns False.
 2. or: Returns True if at least one of the operands is True, returns False only if both are False.
 3. not: Returns the opposite boolean value of the operand. If the operand is True, it returns False, and vice versa.
These operators are commonly used in conditional statements if, elif, else to control the flow of a program based on multiple conditions.


9. What is type casting in Python?
   * Type casting in Python, also known as type conversion, refers to the process of changing a variable's data type to another. It is useful when you need to perform operations that require specific data types or when you want to represent data in a different format. Python supports both implicit and explicit type casting. Implicit type casting is automatic and Explicit type casting is Manual.


10. What is the difference between implicit and explicit type casting?
   1. Implicit type casting is automatic, Python automatically converts one data type to another in certain situations, often to avoid data loss. For example, when you add an integer and a float, Python implicitly converts the integer to a float before performing the addition.

   2. Explicit type casting is manual, Python provides built-in functions that allow you to explicitly convert a variable from one data type to another.


11. What is the purpose of conditional statements in Python?
    * The purpose of conditional statements in Python is to control the flow of execution in a program based on whether certain conditions are met. The primary conditional statements in Python are if, elif, and else. They are fundamental for implementing decision-making logic and creating complex program flows, making them more dynamic and adaptable.


12. How does the elif statement work?
  * The “elif” keyword in Python, stands for “else if”. It can be used in conditional statements to check for multiple conditions. For example, if the first condition is false, it moves on to the next “elif” statement to check if that condition is true.
   


13. What is the difference between for and while loops?
  * In Python, for and while loops are used to repeat a block of code multiple times, but they are suited to different use cases based on how the iteration is controlled.

    1. For loop: Known number of iterations or iterating over a sequence,known number of iterations or iterating over a sequence, best for iterating through collections, terminates when the sequence is exhausted.

    2. While loop: Unknown number of iterations; depends on a condition, controlled by a logical condition, best for looping until a condition changes, terminates when the condition becomes false.


14. Describe a scenario where a while loop is more suitable than a for loop.
    * A while loop is more suitable than a for loop in scenarios where the number of iterations is not predetermined and depends on a condition being met. One common scenario is when waiting for user input to satisfy a condition or when a process must run until a specific state is achieved.

     Scenario: Password validation
     Suppose you are building a system that asks the user to enter a password and continues prompting them until they provide the correct password.
Why Use while?
The number of attempts the user will take is unknown.
The loop should continue until the correct password is entered.

     1. The end condition is not tied to a predetermined count but rather depends on dynamic factors.
     2. The loop needs to adapt based on user input, real-time data, or external events.
     3. A while loop is ideal because it allows for flexible, condition-driven iteration

```python
Q14. original_string = input("15: ")

reversed_string = ""
index = len(original_string) - 1

while index >= 0:
    reversed_string += original_string[index]
    index -= 1

print("Reversed string:", reversed_string)
```

```python
Q15. number = int(input("10: "))

if number < 0:
    print("Factorial is not defined for negative numbers.")
else:
    factorial = 1
    current = number

    while current > 0:
        factorial *= current
        current -= 1

    # Print the result
    print(f"The factorial of {number} is: {factorial}")
```
